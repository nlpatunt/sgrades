```python
import pandas as pd
import csv

def detect_problematic_rows(file_path):
    """Detect problematic rows in CSV file"""
    
    print("🔍 Analyzing q1/test.csv for problematic rows...")
    
    # Method 1: Try to read with pandas and catch errors
    try:
        df = pd.read_csv(file_path)
        print(f"✅ Pandas can read the file: {len(df)} rows")
    except Exception as e:
        print(f"❌ Pandas error: {e}")
        return
    
    # Method 2: Check each row manually
    problematic_rows = []
    
    with open(file_path, 'r', encoding='utf-8') as f:
        reader = csv.reader(f)
        headers = next(reader)
        
        for row_num, row in enumerate(reader, start=2):  # Start from 2 (after header)
            try:
                # Check row length
                if len(row) != len(headers):
                    problematic_rows.append({
                        'row_num': row_num,
                        'issue': f'Wrong column count: {len(row)} vs {len(headers)}',
                        'data': row[:3]  # First 3 columns for reference
                    })
                
                # Check for problematic score columns (should be empty)
                if len(row) >= 12:  # Ensure we have score columns
                    score_cols = row[8:12]  # score_1, score_2, score_3, score_outlier
                    non_empty_scores = [i for i, score in enumerate(score_cols) if score.strip() != '']
                    if non_empty_scores:
                        problematic_rows.append({
                            'row_num': row_num,
                            'issue': f'Non-empty scores in columns: {non_empty_scores}',
                            'data': row[:3]  # First 3 columns for reference
                        })
                
                # Check for very long text that might break parsing
                for col_idx, cell in enumerate(row):
                    if len(cell) > 5000:  # Very long text
                        problematic_rows.append({
                            'row_num': row_num,
                            'issue': f'Very long text in column {col_idx}: {len(cell)} chars',
                            'data': row[:3]
                        })
                
            except Exception as e:
                problematic_rows.append({
                    'row_num': row_num,
                    'issue': f'Parse error: {e}',
                    'data': row[:3] if row else 'No data'
                })
    
    # Report findings
    if problematic_rows:
        print(f"\n❌ Found {len(problematic_rows)} problematic rows:")
        for issue in problematic_rows:
            print(f"Row {issue['row_num']}: {issue['issue']}")
            print(f"  Data preview: {issue['data']}")
            print()
    else:
        print("✅ No obvious issues found")
    
    # Method 3: Check specific columns for data type issues
    print("\n🔍 Checking column data types...")
    for col in df.columns:
        unique_vals = df[col].nunique()
        sample_val = df[col].iloc[0] if len(df) > 0 else 'N/A'
        print(f"{col}: {unique_vals} unique values, sample: {str(sample_val)[:50]}...")
    
    return problematic_rows

# Run the detection
problematic = detect_problematic_rows('/home/ts1506.UNT/Downloads/Temp/grade_like_a_human_dataset_os/q1/test_noLabels.csv')

# Suggest which rows to delete
if problematic:
    print("\n🗑️ ROWS TO DELETE:")
    rows_to_delete = [issue['row_num'] for issue in problematic]
    print(f"Delete rows: {rows_to_delete}")
    print("\nOr run this command to create a clean version:")
    print("df_clean = df.drop(df.index[list_of_bad_indices])")
else:
    print("\n✅ File looks clean - the issue might be elsewhere")
```

```python
import pandas as pd
import os

# Your actual directory path
base_dir = '/home/ts1506.UNT/Downloads/Temp/grade_like_a_human_dataset_os'
q1_dir = f'{base_dir}/q1'
q2_dir = f'{base_dir}/q2'

def analyze_and_fix_q1():
    """Analyze q1 files and fix any issues"""
    
    print(f"🔍 Analyzing files in: {q1_dir}")
    
    # Check what files exist in q1 and q2
    q1_files = os.listdir(q1_dir) if os.path.exists(q1_dir) else []
    q2_files = os.listdir(q2_dir) if os.path.exists(q2_dir) else []
    
    print(f"q1 files: {q1_files}")
    print(f"q2 files: {q2_files}")
    
    # Check each required file
    required_files = ['train.csv', 'validation.csv', 'test_noLabels.csv']
    
    for file in required_files:
        q1_file = f'{q1_dir}/{file}'
        q2_file = f'{q2_dir}/{file}'
        
        print(f"\n📋 Analyzing {file}:")
        
        # Check if files exist
        q1_exists = os.path.exists(q1_file)
        q2_exists = os.path.exists(q2_file)
        
        print(f"  q1/{file} exists: {q1_exists}")
        print(f"  q2/{file} exists: {q2_exists}")
        
        if q1_exists:
            # Analyze q1 file
            try:
                df_q1 = pd.read_csv(q1_file)
                print(f"  q1/{file}: {len(df_q1)} rows, {len(df_q1.columns)} columns")
                
                # Check for empty file
                if len(df_q1) == 0:
                    print(f"  ❌ {file} is EMPTY! This is the problem.")
                    
                    # Try to fix with q2 data if available
                    if q2_exists:
                        print(f"  🔧 Copying {file} from q2 to fix empty q1 file...")
                        df_q2 = pd.read_csv(q2_file)
                        
                        # For test files, clear score columns
                        if file == 'test_noLabels.csv':
                            df_q2['score_1'] = ''
                            df_q2['score_2'] = ''
                            df_q2['score_3'] = ''
                            df_q2['score_outlier'] = ''
                        
                        # Save fixed file
                        df_q2.to_csv(q1_file, index=False)
                        print(f"  ✅ Fixed {q1_file} with {len(df_q2)} rows")
                
            except Exception as e:
                print(f"  ❌ Error reading q1/{file}: {e}")
                
                # Try to fix with q2
                if q2_exists:
                    print(f"  🔧 Replacing corrupted {file} with q2 version...")
                    df_q2 = pd.read_csv(q2_file)
                    
                    # For test files, clear score columns
                    if file == 'test_noLabels.csv':
                        df_q2['score_1'] = ''
                        df_q2['score_2'] = ''
                        df_q2['score_3'] = ''
                        df_q2['score_outlier'] = ''
                    
                    df_q2.to_csv(q1_file, index=False)
                    print(f"  ✅ Replaced {q1_file} with {len(df_q2)} rows")
        
        elif not q1_exists and q2_exists:
            # q1 file missing, copy from q2
            print(f"  🔧 Creating missing {q1_file} from q2...")
            df_q2 = pd.read_csv(q2_file)
            
            # For test files, clear score columns
            if file == 'test_noLabels.csv':
                df_q2['score_1'] = ''
                df_q2['score_2'] = ''
                df_q2['score_3'] = ''
                df_q2['score_outlier'] = ''
            
            df_q2.to_csv(q1_file, index=False)
            print(f"  ✅ Created {q1_file} with {len(df_q2)} rows")

# Run the analysis and fix
analyze_and_fix_q1()

print(f"\n🎯 Summary:")
print(f"After running this script, all q1 files should be fixed.")
print(f"Next steps:")
print(f"1. Upload the fixed files from {q1_dir} to HuggingFace")
print(f"2. Restart your BESESR server")
print(f"3. Check if grade_like_a_human_dataset_os_q1 now loads")
```

```python
import pandas as pd
import os

# Source directory
source_dir = '/home/ts1506.UNT/Downloads/Temp/grade_like_a_human_dataset_os/q1'
# Target directory (nested q1/q1/)
target_dir = '/home/ts1506.UNT/Downloads/Temp/grade_like_a_human_dataset_os/q1/q1'

def recreate_q1_splits():
    """Recreate q1 splits with proper test data"""
    
    print(f"🔧 Creating new q1 splits...")
    print(f"Source: {source_dir}")
    print(f"Target: {target_dir}")
    
    # Create target directory
    os.makedirs(target_dir, exist_ok=True)
    
    # Read original files
    train_file = f'{source_dir}/train.csv'
    val_file = f'{source_dir}/validation.csv'
    
    print(f"\n📖 Reading source files...")
    
    # Read train and validation
    df_train = pd.read_csv(train_file)
    df_val = pd.read_csv(val_file)
    
    print(f"Original train: {len(df_train)} rows")
    print(f"Original validation: {len(df_val)} rows")
    
    # Take last 2 rows from train for test
    test_from_train = df_train.tail(2).copy()
    new_train = df_train.iloc[:-2].copy()  # Remove last 2 rows
    
    # Take last 1 row from validation for test
    test_from_val = df_val.tail(1).copy()
    new_val = df_val.iloc[:-1].copy()  # Remove last 1 row
    
    # Combine test data (2 from train + 1 from val = 3 test rows)
    df_test = pd.concat([test_from_train, test_from_val], ignore_index=True)
    
    print(f"\n📊 New splits:")
    print(f"New train: {len(new_train)} rows")
    print(f"New validation: {len(new_val)} rows")
    print(f"New test: {len(df_test)} rows")
    
    # Clear score columns in test data (make it "without labels")
    df_test['score_1'] = ''
    df_test['score_2'] = ''
    df_test['score_3'] = ''
    df_test['score_outlier'] = ''
    
    print(f"\n✅ Cleared score columns in test data")
    
    # Save new files in target directory
    new_train_file = f'{target_dir}/train.csv'
    new_val_file = f'{target_dir}/validation.csv'
    new_test_file = f'{target_dir}/test.csv'
    
    new_train.to_csv(new_train_file, index=False)
    new_val.to_csv(new_val_file, index=False)
    df_test.to_csv(new_test_file, index=False)
    
    print(f"\n💾 Saved files:")
    print(f"✅ {new_train_file}")
    print(f"✅ {new_val_file}")
    print(f"✅ {new_test_file}")
    
    # Verify the files
    print(f"\n🔍 Verification:")
    for file_path in [new_train_file, new_val_file, new_test_file]:
        if os.path.exists(file_path):
            df_check = pd.read_csv(file_path)
            file_name = os.path.basename(file_path)
            print(f"{file_name}: {len(df_check)} rows, {len(df_check.columns)} columns")
            
            # Check score columns for test file
            if file_name == 'test.csv':
                score_empty = all(df_check[col].isna().all() or (df_check[col] == '').all() 
                                for col in ['score_1', 'score_2', 'score_3', 'score_outlier'])
                print(f"  Score columns empty: {score_empty}")
        else:
            print(f"❌ File not created: {file_path}")
    
    # Show sample data
    print(f"\n📋 Sample test data (without scores):")
    print(df_test[['question_id', 'id', 'answer', 'score_1', 'score_2']].head())
    
    return new_train_file, new_val_file, new_test_file

# Run the recreation
created_files = recreate_q1_splits()

print(f"\n🎯 Next Steps:")
print(f"1. Upload the files from {target_dir}/ to HuggingFace q1/ folder")
print(f"2. Replace the existing q1/train.csv, q1/validation.csv, q1/test.csv")
print(f"3. Restart your BESESR server")
print(f"4. Check if grade_like_a_human_dataset_os_q1 loads successfully")
```
